```python
import pandas as pd
```

```python
orders = pd.read_csv('../datasets/raw/olist_orders_dataset.csv')
orders.head()
```

```python
orders.count()
```

```python
items = pd.read_csv('../datasets/raw/olist_order_items_dataset.csv')
items.head()
```

```python
items.count()
```

```python
joinned_order_items = pd.merge(orders, items, on='order_id')
joinned_order_items.head()
```

```python
joinned_order_items.count()
```

```python
products = pd.read_csv('../datasets/raw/olist_products_dataset.csv')
products.head()
```

```python
joinned_products = pd.merge(joinned_order_items, products, on='product_id')
joinned_products.head()
```

```python
joinned_products.count()
```

```python
payments = pd.read_csv('../datasets/raw/olist_order_payments_dataset.csv')
payments.head()
```

```python
payments.count()
```

```python
joinned_payments = pd.merge(joinned_products, payments, on='order_id')
joinned_payments.head()
```

```python
joinned_payments.count()
```

```python
reviews = pd.read_csv('../datasets/raw/olist_order_reviews_dataset.csv')
reviews.head()
```

```python
reviews.count()
```

```python
joinned_reviews = pd.merge(joinned_payments, reviews, on='order_id')
joinned_reviews.head()
```

```python
joinned_reviews.count()
```

```python
customers = pd.read_csv('../datasets/raw/olist_customers_dataset.csv')
customers.head()
```

```python
customers.count()
```

```python
data = pd.merge(joinned_reviews, customers, on='customer_id')
data.head()
```

```python
data.count()
```
